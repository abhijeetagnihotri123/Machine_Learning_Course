```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Amazon - Movies and TV Ratings.csv')
df
```

```python
#part 1 of the question
#creating another dataframe and replacing NaN with dummy value -2
#will ease in comparison
df1 = pd.DataFrame()
df1 = df.fillna(-2)
df1.drop('user_id',axis='columns',inplace = True)
df1
```

```python
c = 0
max_views = 0
for x in df1:
    c=0
    for val in df1[x]:
        if val != -2:
            c+=1
    if max_views < c:
        max_views = c
```

```python
l = []
for x in df1:
    c = 0
    for val in df1[x]:
        if val != -2:
            c+=1
    if c == max_views:
        l.append(x)
#l holds the list of movies which has maximum rating
l
```

```python
#part 2 of the question
#the list l will be holding all the average rating for the movie# respectively
l1 = []
l = []
movies = []
average_ratings = []
for x in df1:
    for val in df1[x]:
        if val != -2:
            l1.append(val)
    a = np.array(l1)
    y = a.mean()
    movies.append((y,x))
    l.append(y)
average_ratings = l #holds all the rating options for that movie# respectively
movies.sort()
print(movies[-1],movies[-2],movies[-3],movies[-4],movies[-5])
```

```python
import matplotlib.pyplot as plt
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Movie ratings with every movie number')
arr = np.array(l)
plt.hist(arr,bins=11)
```

```python
#part 3 of question
c = 1
f = 0
l = []
for x in df1:
    f=0
    for val in df1[x]:
        if val != -2:
            f+=1
    t = (f,x)
    l.append(t)
    c+=1
l.sort()
print("Top 5 movies with least audience in the form of a tuple")
print("First number represents audience number")
print("second number represents movie number")
l[:5]
```

```python
#part 4 of the question
#To build a recommendation system we use Logistics regression
#also we use average ratings of the films to divide the movies to recommended or not recommended
#We set thresold of the recommended as the medial of average ratings 
#if rating is below the thresold then 0 otherwise 1
```

```python
threshold_rating = np.mean(average_ratings)
threshold_rating
```

```python
l = []
#everything is numbered in accordance to movie number
#l holds whether the respective movie number is recommended or not 0 for no 1 for yes
for x in average_ratings:
    if x >= threshold_rating:
        l.append(1)
    else:
        l.append(0)
```

```python
X = pd.DataFrame()
y = pd.DataFrame()
```

```python
X['Average_ratings']=average_ratings
y['recommended']=l
```

```python
#part 4 we take 25% for testing 75% for training
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
len(X_train),len(X_test),len(y_train),len(y_test)
```

```python
#part 5 we use simple logistics regression model
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg
```

```python
reg.fit(X_train,y_train)
```

```python
#part 6 noting the score this approach is to be verified from instructor
reg.score(X_test,y_test)
```
