```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('HR_comma_sep.csv')
df
```

```python
df.Department.unique()
```

```python
l = []
for val in df.Department:
    if val == 'sales':
        l.append(0)
    elif val == 'accounting':
        l.append(1)
    elif val == 'hr':
        l.append(2)
    elif val == 'technical':
        l.append(3)
    elif val == 'support':
        l.append(4)
    elif val == 'management':
        l.append(5)
    elif val == 'IT':
        l.append(6)
    elif val == 'product_mng':
        l.append(7)
    elif val == 'marketing':
        l.append(8)
    else:
        l.append(9)
df['Department'] = l
df
```

```python
df.salary.unique()
```

```python
l = []
for val in df.salary:
    if val == 'low':
        l.append(0)
    elif val == 'medium':
        l.append(1)
    else:
        l.append(2)
df['salary'] = l
```

```python
m = df['average_montly_hours'].mean()
```

```python
l = []
for val in df.average_montly_hours:
    l.append(val/m)
df['average_montly_hours'] = l
```

```python
X = df.drop(['left'],axis='columns')
y = df.left
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
len(X_train)
```

```python
reg = LogisticRegression()
reg
```

```python
reg.fit(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```
